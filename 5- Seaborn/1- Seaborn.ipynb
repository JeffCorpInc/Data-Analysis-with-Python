```python
## IMPORTING SEABORN

import seaborn as sns

## BASIC PLOTTING

data = sns.load_dataset('tips')
```

```python
## SCATTER PLOT

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=data)
plt.title('Scatter Plot for Total Bills vs Tips')
plt.show()
```

```python
## LINE PLOT

sns.lineplot(x='size',y='total_bill',data=data)
plt.title('Line Plot of Total bill paid by How many People')
plt.show()
```

```python
## CATEGORICAL PLOT


sns.barplot(x='day',y='total_bill',data=data)
plt.title('Catrgorical chart of Total bill vs Day')
plt.show()
```

```python
## BOX PLOT

sns.boxplot(x='day',y='total_bill',data=data)
plt.title('Box Plot')
plt.show()
```

```python
## VOILIN PLOT

sns.violinplot(x='day',y='total_bill',data=data)
plt.title('Voilin Plot')
plt.show()
```

```python
## HISTOGRAM

sns.histplot(data['total_bill'],bins=10,kde=True)                # kde = will show curve
plt.title('Histogram')
plt.show()
```

```python
## KDE PLOT

sns.kdeplot(data['total_bill'],fill=True)
plt.title('KDE Plot')
plt.show()
```

```python
## PAIR PLOT

sns.pairplot(data)                          # This will show all the numerical relations between plots
```

```python
# Correlation

corr = data[['total_bill','tip','size']].corr()
corr
```

```python
## HEATMAP

sns.heatmap(corr,annot=True,cmap='coolwarm')
```
